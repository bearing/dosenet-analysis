```python
def graph_dosenet(location_name):
    location1_rawdata = get_url(location_name)
    devicetime = location1_rawdata['deviceTime_utc']
    pm1 = location1_rawdata['PM1']
    pm25 = location1_rawdata['PM25']
    pm10 = location1_rawdata['PM10']

    fig = go.Figure()
    fig.add_trace(go.Scatter(
            x = devicetime,
            y = pm1,
            name = 'pm1'))
    fig.add_trace(go.Scatter(
            x = devicetime,
            y = pm25,
            name = 'pm25'))

    fig.add_trace(go.Scatter(
            x = devicetime,
            y = pm10,
            name = 'pm10'))

    fig.update_layout(plot_bgcolor='white',width=1000, height=450, title = 'Line chart')
    fig.update_yaxes(title="air quality",titlefont=dict(color='black', size=20),
                    showgrid=False,tickcolor='black',
                    tickfont=dict(color='black', size=16))
    fig.update_xaxes(title="Time (local) ",titlefont=dict(color='black', size=20),
                    linecolor='black',tickfont=dict(color='black',size=12))
    fig.update_layout(legend_orientation="h",
                    legend=dict(x=0,y=-.2, font=dict(size=13)))
    fig.update_layout(barmode='group')
    
    fig
    return fig
```

```python
def graph_histogram(file_name, 
               file_location = r'C:\Users\Richie Woo\Documents\python-scripts\processed_data',
               save = False,
               show = True):
    all_data = pd.read_csv(file_name)
    data = all_data['pm25diff']
    dataavg = np.mean(data)
    
    fig = go.Figure()
    
    fig.add_trace(go.Histogram(x=data, name = 'Trace 1'))
    
    fig.update_layout(barmode='overlay')
    fig.update_traces(opacity=0.75)
    
    fig.update_layout(plot_bgcolor='white',width=1000, height=450, title = str('%s, Histogram, AVG = %s' % (file_name, dataavg)))
    fig.update_yaxes(title="Count",titlefont=dict(color='black', size=20),
                     showgrid=False,tickcolor='black',
                     tickfont=dict(color='black', size=16))
    fig.update_xaxes(title="PM 2.5 ug/mL Avg Difference / Hour",titlefont=dict(color='black', size=20),
                     linecolor='black',tickfont=dict(color='black',size=12))
    fig.update_layout(legend_orientation="h",
                      legend=dict(x=0,y=-.2, font=dict(size=13)))
    
    if show == True:
        fig.show()
    if save == True:
        os.chdir(file_location)
        fig.write_html(str(file_name[0:-4] + '_histogram.html'))
        
    os.chdir(os.path.dirname(os.getcwd()))
    
    return fig
```

```python
def graph_correlation(file_name, 
               file_location = r'C:\Users\Richie Woo\Documents\python-scripts\processed_data',
               save = False,
               show = True):
    all_data = pd.read_csv(file_name)
    
    new_file_name = file_name[:-4]
    print(new_file_name)
    
    pm25d = all_data['PM25d']
    pm25p = all_data['PM25p']
    
    pcorr, _ = pearsonr(pm25d, pm25p)
    pcorr = round(pcorr, 3)
    spcorr, _ = spearmanr(pm25d, pm25p)
    spcorr = round(spcorr, 3)
    
    titlestring = str('%s, Dosenet VS Purpleair, Pearson''s r = %s, Spearman''s r = %s' %(new_file_name, pcorr, spcorr))
    
    fig = go.Figure()
    
    fig.add_trace(go.Scatter(
            x = pm25d,
            y = pm25p,
            name = 'Dosenet VS Purpleair',
            mode = 'markers'))


    fig.update_layout(plot_bgcolor='white',width=1000, height=450, title = titlestring)
    fig.update_yaxes(title="Purpleair Air Quality",titlefont=dict(color='black', size=20),
                    showgrid=False,tickcolor='black',
                    tickfont=dict(color='black', size=16))
    fig.update_xaxes(title="Dosenet Air Quality",titlefont=dict(color='black', size=20),
                    linecolor='black',tickfont=dict(color='black',size=12))
    fig.update_layout(legend_orientation="h",
                    legend=dict(x=0,y=-.2, font=dict(size=13)))
    fig.update_layout(barmode='group')
    
    
    if show == True:
        fig.show()
    if save == True:
        os.chdir(file_location)
        fig.write_html(str(new_file_name + '_diffgraph.html'))
        
    os.chdir(os.path.dirname(os.getcwd()))
    return fig
```

```python
def get_url(location_name):
    '''Gets the csv location data from radwatch downloads. '''
    
    url = 'https://radwatch.berkeley.edu/test/dosenet/' + location_name + '.csv'

    header = {
      "User-Agent": "Mozilla/5.0 (X11; Linux x86_64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/50.0.2661.75 Safari/537.36",
      "X-Requested-With": "XMLHttpRequest"
    }
    
    s=requests.get(url,headers=header).text
    raw_cpm_data = pd.read_csv(io.StringIO(s))
    
    return raw_cpm_data
```
